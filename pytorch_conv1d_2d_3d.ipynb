## conv1d

```python
# input shape: (N, Cin, Lin)
# filter shape: (out_channerls, in_channles/group, kW)
# output shape: (N, Cout=out_channels, Lout)
```

```python
import torch
```

```python
inputs = torch.ones(5, 16, 50)
```

```python
filters = torch.ones(2, 16, 1)
```

```python
import torch.nn.functional as F
```

```python
F.conv1d(inputs, filters).shape
```

```python
inputs = torch.ones(2, 3, 4)
```

```python
filters = torch.ones(1, 3, 3)
```

```python
F.conv1d(inputs, filters)
```

## conv2d

```python
# input shape: (N, Cin, Hin, Win)
# filter shape: (out_channerls, in_channles/group, kH, kW)
# output shape: (N, Cout, Hout, Wout)
```

```python
inputs = torch.randn(5, 3, 28, 28)
```

```python
filters = torch.randn(2, 3, 2, 2)
```

```python
F.conv2d(inputs, filters).shape
```

```python
inputs = torch.ones(5, 3, 28, 28)
```

```python
filters = torch.ones(2, 3, 2, 2)
```

```python
F.conv2d(inputs, filters)
```

## conv3d

```python
# input shape: (N, Cin, Din, Hin, Win)
# filter shape: (out_channerls, in_channles/group, kD, kH, kW)
# output shape: (N, Cout, Dout, Hout, Wout)
```

```python
inputs = torch.randn(5, 3, 28, 28, 28)
```

```python
filters = torch.randn(2, 3, 2,2, 2)
```

```python
F.conv3d(inputs, filters).shape
```

```python
inputs = torch.ones(5, 3, 28, 28, 28)
```

```python
filters = torch.ones(2, 3, 2,2, 2)
```

```python
F.conv3d(inputs, filters)
```
